```python
# authors with the most book 
df_grouping = df.groupby(['Author']).count().sort_values(by='Score', ascending=False)
df_grouping.head(10)
```

```python
# visualization authors with the highest amount of books
df_grouping.head(10).plot(kind='bar', y='Title', color='red', ylabel='Title', xlabel='Author', title = 'Best 10 Authors based on the count of books', stacked=True)
```

```python
# top 10 best authors based on average rating
df_grouping = df.groupby(['Author']).mean('Rating').sort_values(by='Rating', ascending=False)
df_grouping.head(10)
```
